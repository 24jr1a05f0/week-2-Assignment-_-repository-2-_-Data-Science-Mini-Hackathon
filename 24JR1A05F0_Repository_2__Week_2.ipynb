```python
import pandas as pd

# Create dataset
df = pd.DataFrame({
    "Category": ["A","B","A","C","B","A","C"],
    "Value1": [10,20,30,40,50,60,70],
    "Value2": [100,200,300,400,500,600,700]
})

# Save file
df.to_csv("data.csv", index=False)

print("✅ data.csv created")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------
# DATASET INTRODUCTION
# ---------------------------
print("📊 DATASET INTRODUCTION")
print("This dataset contains Category, Value1, and Value2 columns.\n")

# Load dataset
df = pd.read_csv("data.csv")

print("Shape:", df.shape)
print("Columns:", df.columns)
print(df.head())

# ---------------------------
# DATA CLEANING
# ---------------------------
df.drop_duplicates(inplace=True)
df.fillna(method='ffill', inplace=True)

# ---------------------------
# EXPLORATORY DATA ANALYSIS
# ---------------------------
print("\n📈 EDA")
print(df.describe())

# ---------------------------
# FILTER (INTERACTIVE)
# ---------------------------
print("\nAvailable Categories:", df["Category"].unique())
choice = input("Enter Category (A/B/C): ")

filtered_df = df[df["Category"] == choice]

# ---------------------------
# KPI METRICS
# ---------------------------
print("\n📊 KPIs")
print("Total Rows:", len(filtered_df))
print("Average Value1:", filtered_df["Value1"].mean())
print("Average Value2:", filtered_df["Value2"].mean())

# ---------------------------
# VISUALIZATIONS
# ---------------------------

# Histogram
plt.figure()
plt.hist(filtered_df["Value1"])
plt.title("Distribution of Value1")
plt.show()

# Scatter Plot
plt.figure()
plt.scatter(filtered_df["Value1"], filtered_df["Value2"])
plt.title("Value1 vs Value2")
plt.show()

# Bar Chart
plt.figure()
df["Category"].value_counts().plot(kind='bar')
plt.title("Category Distribution")
plt.show()

# Heatmap
plt.figure()
sns.heatmap(df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()

# ---------------------------
# INSIGHTS
# ---------------------------
print("\n🔍 INSIGHTS")
print("- Value1 and Value2 increase together (positive correlation)")
print("- Category A appears more frequently")
print("- Higher values are seen in later entries")

# ---------------------------
# CONCLUSION
# ---------------------------
print("\n✅ CONCLUSION")
print("The dataset shows a clear relationship between Value1 and Value2.")
print("Category-wise filtering helps understand patterns better.")
```
